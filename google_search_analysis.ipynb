```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from pytrends.request import TrendReq
```

```python
?TrendReq
```

```python
trends=TrendReq()
```

```python
trends.build_payload(kw_list=["Data Science"],timeframe='2019-10-01 2024-01-01')
data = trends.interest_by_region()
data = data.sort_values(by='Data Science', ascending= False)
data = data.head(10)
print(data)
```

```python
data.reset_index().plot(x='geoName',y='Data Science',figsize=(20,15),kind='bar')
plt.style.use('fivethirtyeight')
plt.show()
```

```python
trends.build_payload(kw_list=["Data Science"],timeframe= '2019-10-01 2024-01-01' , geo='IN', gprop = 'images')
data = trends.interest_by_region()
data = data.sort_values(by='Data Science', ascending= False)
data = data.head(10)
print(data)
```

```python
data.reset_index().plot(x='geoName',y='Data Science',figsize=(20,15),kind='bar')
plt.style.use('fivethirtyeight')
plt.show()
```

```python
trends.build_payload(kw_list=["Data Science"],timeframe= '2019-10-01 2024-01-01' , geo='IN', gprop = 'youtube')
data = trends.interest_by_region()
data = data.sort_values(by='Data Science', ascending= False)
data = data.head(10)
print(data)
```

```python
data.reset_index().plot(x='geoName',y='Data Science',figsize=(15,10),kind='bar')
plt.style.use('/content/custom.mplstyle')
plt.show()
```

```python
data = TrendReq(hl='en-US',tz=360)
data.build_payload(kw_list=['Data Science'])
data=data.interest_over_time()
fig, ax = plt.subplots(figsize=(30,20))
data['Data Science'].plot()
plt.style.use('fivethirtyeight')
plt.title('Total Google Searches for Data Science',fontweight='bold')
plt.xlabel('Year')
plt.ylabel('Total Count')
plt.show()
```
